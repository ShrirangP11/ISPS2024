```python
import xarray as xr
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from datetime import date as date
from pyextremes import EVA
from pyextremes.plotting import plot_extremes
from pyextremes import get_extremes
import seaborn as sns
import scipy
```

```python
#Extracting daily rainfall(mm) values over Pune for years 1951-2023
lat_range = slice(16.0, 19.0)
lon_range = slice(73.0, 76.0)
daily_rainfall = np.array([]) #Create empty array
for year in range(1951,2024):
    dataset = xr.open_dataset(f"C:\\Users\\ShrirangP\\Downloads\\Rainfall_IMD_025_NETCDF_1951-2023\\RF25_ind{year}_rfp25.nc")
    pune = dataset.sel(LATITUDE=lat_range, LONGITUDE=lon_range)
    #daily_rainfall = np.append(daily_rainfall, (pune['RAINFALL'].to_series()).dropna())
    dates = pd.date_range(start=f"{year}-01-01",end=f"{year}-12-31")
    for date in dates:
        df = pune['RAINFALL'].sel(TIME=date)
        val = (df.to_series()).dropna()
        daily_rainfall = np.append(daily_rainfall, val.max())
```

```python
#Extracting maximum yearly rainfall for years 1951-2023
lat_range = slice(16.0, 19.0)
lon_range = slice(73.0, 76.0)
max_rainfall = np.array([]) #Create empty array
for year in range(1951,2024):
    dataset = xr.open_dataset(f"C:\\Users\\ShrirangP\\Downloads\\Rainfall_IMD_025_NETCDF_1951-2023\\RF25_ind{year}_rfp25.nc")
    pune = dataset.sel(LATITUDE=lat_range, LONGITUDE=lon_range)
    max_rainfall = np.append(max_rainfall, (pune['RAINFALL'].to_series()).max())
```

```python
data = pd.Series(max_rainfall)
x = np.arange(0,73)
p=sns.regplot(x=x,y=data)
sns.lineplot(x=x, y=data)
p.set(xlabel='Year Index', ylabel='Maximum Rainfall(mm)')
slope, intercept, r, p, sterr = scipy.stats.linregress(x=p.get_lines()[0].get_xdata(),
                                                       y=p.get_lines()[0].get_ydata())

#add regression equation to plot
plt.text(2, 600, 'y = ' + str(round(intercept,3)) + ' + ' + str(round(slope,3)) + 'x')
```

```python
#Replacing NaN values with zero
lat_range = slice(10.0, 19.0)
lon_range = slice(73.0, 76.0)
values = np.array([]) #Create empty array
for year in range(1980,2023):
    dataset = xr.open_dataset(f"C:\\Users\\ShrirangP\\Downloads\\Rainfall_IMD_025_NETCDF_1951-2023\\RF25_ind{year}_rfp25.nc")
    pune_rf = dataset.sel(LATITUDE=lat_range, LONGITUDE=lon_range)
    sliced = pune_rf['RAINFALL'].sel(TIME=slice(f"{year}-06-01", f"{year}-08-31"))
    rain = sliced.to_series()
    rain = rain.fillna(0)
    days = int(rain.count()/481) #Number of days
    bsize = 481
    averages = np.zeros(days)
    for i in range(days):
        averages[i] = sum(rain.values[i*bsize:(i+1)*bsize])/bsize
    values = np.append(values,averages)
    
print(values.shape)
```

```python
y = values.round(3)
x = np.arange(1,values.shape[0]+1)
ynormed = (y - y.mean())/(y.std())
plt.plot(x,ynormed)
```

```python
#Removing NaN values
lat_range = slice(15.0, 25.0)
lon_range = slice(70.0, 90.0)
values2 = np.array([]) #Create empty array
for year in range(1980,2020):
    dataset = xr.open_dataset(f"C:\\Users\\ShrirangP\\Downloads\\Rainfall_IMD_025_NETCDF_1951-2023\\RF25_ind{year}_rfp25.nc")
    pune_rf = dataset.sel(LATITUDE=lat_range, LONGITUDE=lon_range)
    t = np.array([])
    dates = pd.date_range(start=f"{year}-06-01",end=f"{year}-08-31")
    for date in dates:
        df = pune_rf['RAINFALL'].sel(TIME=date)
        val = df.to_series()
        val = val.dropna()
        t = np.append(t,val.mean())
#     print(t.shape)
    standardized = (t-np.mean(t))/np.std(t)
    values2 = np.append(values2,standardized)
total_days = values2.shape[0]
print(total_days)
```

```python
#Removing NaN values
lat_range = slice(15.0, 25.0)
lon_range = slice(70.0, 90.0)
values2 = np.array([]) #Create empty array
std = np.array([])
for year in range(1951,2008):
    dataset = xr.open_dataset(f"C:\\Users\\ShrirangP\\Downloads\\Rainfall_IMD_025_NETCDF_1951-2023\\RF25_ind{year}_rfp25.nc")
    pune_rf = dataset.sel(LATITUDE=lat_range, LONGITUDE=lon_range)
    t = np.array([])
    dates = pd.date_range(start=f"{year}-07-01",end=f"{year}-08-31")
    for date in dates:
        df = pune_rf['RAINFALL'].sel(TIME=date)
        val = df.to_series()
        val = val.dropna()
        t = np.append(t,val.mean())
#     print(t.shape)
    std = np.append(std,np.std(t))
    values2 = np.append(values2,t)
total_days = values2.shape[0]
print(total_days)
```

```python
new_std=np.zeros(0)
for i in range(int(total_days/dates.shape[0])):
    new_std = np.append(new_std,np.repeat(std[i],92))
    
```

```python
global_mean = np.mean(values2)
normalized = np.zeros(total_days)
for i in range(total_days):
    normalized[i] = (values2[i]-global_mean)/new_std[i]
```
