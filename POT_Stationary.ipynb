```python
import pandas as pd
from pyextremes import EVA
import matplotlib.pyplot as plt
from pyextremes import plot_mean_residual_life
```

```python
rf_data = pd.read_csv(r"C:\Users\admin\Desktop\ISPS\daily_rainfall.csv", header = None)
rf_values = rf_data.iloc[0].tolist()
date_range = pd.date_range(start='1951-01-01', end='2023-12-31', freq='D')
rf_datetime = pd.Series(rf_values, index = date_range)
```

```python
model = EVA(rf_datetime)
model.get_extremes(method="POT",threshold = 204.5, r = "24h")
```

```python
model.plot_extremes()
plt.show()
```

```python
model.fit_model()
model.plot_diagnostic(alpha=0.95)

# Display the plots
plt.show()
```
